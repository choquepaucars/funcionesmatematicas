# Funciones algebraicas

### Función lineal

Tiene la forma de $$f(x)=mx + b$$ donde $m$ y $b$  $\in R$. 

$m$ puede ser calculada por: $$m=\frac{y_{2}-y_{1}}{x_{2}-x_{1}}$$

y $b$ es el punto de corte con el eje $y$. Su dominio es $Dom_{f} = (-\infty, \infty)$. Su imagen es $Im_{f} = (-\infty, \infty)$




```python
N=100   #constante N

m=-1 #pendiente
b=5 # b

def f(x):
  return m*x + b

x=np.linspace(start=-10, stop=10, num=N) #definimos el dominio de nuestra función. variable independiente

y=f(x) #variable dependiente
```

```python
fig, (axes1) = plt.subplots(nrows=1, ncols=1) #creamos el entrono de graficas
axes1.plot(x,y)
axes1.grid() # agrega la cuadricula=grid
plt.show()
```

### Funciones polinómicas

Tiene la forma de $$P(x)=a_{n}x^{n} + a_{n-1}x^{n-1}+...+a_{2}x^{2}+a_{1}x + a_{1}$$

a una función que tiene esta forma se le llama polinomio de grado $n$. A los elementos $a$ los llamaremos coeficientes donde $a \in R$. 

**Por ejemplo:**

$$P(x)= 2x^{7} - x^{4} + 3x^{2} + 4$$

que es un polinomio de grado 7.


### Funciones potencia

Hay unas funciones que son un caso particular de las funciones polinómicas que son las funciones potencia, las cuales tienen la forma:


$$f(x)= x^{a}, a \in R$$ 


**Por ejemplo:**

$$f(x)= x^{2}$$

El dominio de $f(x)=x^{2}$ es $Dom_{f} = (-\infty, \infty)$. Su imagen es $Im_{f} = [0, \infty)$





```python
# Nota el x que usaremos es la que definimos en el codigo de arriba
def f(x):
  return (2*x**7) -(x**4) + (3*x**2) + 4

y=f(x)
y.shape

# Nota: los polinomios de grado para tienen la forma de una "u"
# Nota: los polinomios de grado impar tienen la forma de un "parte superior de una s"
```

```python
fig, (axes1)= plt.subplots(nrows=1, ncols=1)
axes1.plot(x,y)
plt.grid() #agrega cuadritos
plt.show()
```

# Funciones trascendentes

Son funciones que no pueden ser expresadas con polinomios. 
Las funciones trascendentes elementales son las exponenciales, las logarítmicas, las trigonométricas, las funciones trigonométricas inversas, las hiperbólicas y las hiperbólicas inversas.\
Es decir, son aquellas que no pueden ser expresadas mediante un polinomio, un cociente de polinomios o raíces de polinomios.

Algunos ejemplos son las funciones $cos(x)$, $sen(x)$ y $tan(x)$

```python
# x ya esta definida en los primeros codigos

#definición
from turtle import width


def cos(x):
    return np.cos(x)
cos=cos(x) # variable dependiente

def sin(x):
    return np.sin(x)
sin=sin(x) # Variable dependiente

def tan(x):
    return np.tan(x)
tan=tan(x)

#grafico
fig, (axes1,axes2, axes3)=plt.subplots(nrows=1, ncols=3, figsize=(15,3)) # parametro figsize(ancho, altura) existe
axes1.plot(x,cos, "b")
axes1.set_xlabel(xlabel="x")
axes1.set_ylabel(ylabel="cos(x)")

axes2.plot(x,sin, "r")
axes2.set_xlabel(xlabel="x")
axes2.set_ylabel(ylabel="sin(x)")

axes3.plot(x,tan, "y")
axes3.set_xlabel(xlabel="x")
axes3.set_ylabel(ylabel="tan(x)")


fig.tight_layout() #mejora la visivilidad
plt.show()
```

### Función exponencial

Tienen la forma de $$f(x)=a^x$$ donde la base $a$ es una constante positiva. Un gran ejemplo de una función exponencial es usando la base como el número de euler:

$$f(x)=e^x$$

```python

def f(x):
  return np.exp(x) #e^x
y=f(x)

fig, (axes1)=plt.subplots(nrows=1, ncols=1, figsize=(5,3)) #figsize(alto, ancho)
axes1.plot(x,y, color="blue", label="exp(x)") # el parametro label define lo que se pondra en la legenda
axes1.set_title(label="Función exponencial")
axes1.set_xlabel(xlabel="x")
axes1.legend()
plt.show()
```

### Función logaritmo

El logaritmo está definido por la **relación**:

$$log_{b}(x) = n \Longleftrightarrow x=b^n$$ 

donde: 



*   $b$ es la base.
*   $n$ es el exponente al que está elevado la base.
*   $x$ es el resultado de elevar la base $b$ al exponente $n$

Nota: la base es mayor a 0 y diferente a 1(b>0 & b=1)\
Nota: el argumento del logaritmo tiene que ser mayor a 0(x>0)\
Nota: el exponente puede tomar cualquier valor real(n pertenece a R)\
**Ejemplo:**

Teniendo b=2 y n=8, entonces:

$$2^8=256$$

Por lo que $x=256$. Calculando el logaritmo base 2 de $x$ es:

$$log_{2}(256) = 8$$

Nota: Los logaritmos de cierto base nos permiten y son super utiles para conocer un cierto exponente




Exponente al que hay que elevar un número, llamado base, para obtener otro número determinado. Un logaritmo busca el exponente y  de una base a que se ha empleado para llegar a un determinado resultado x.

```python
x=np.linspace(start=0.001, stop=256, num=1000) #nota: x >0
# las computadoras procesan datos discretos porque tinen memoria y siempre se acaba

def f(x):
  return np.log2(x)

y=f(x)

# grafico
fig, (axes1)=plt.subplots(nrows=1, ncols=1, figsize=(12,4))
axes1.set_title(label="Función logaritmica")
axes1.plot(x, y, label="log2(x)")
axes1.set_xlabel(xlabel="x")
axes1.set_ylabel(ylabel="f(x)")
axes1.legend()
plt.show()
```

# Función seccionada

Son funciones que tienen diferentes valores definidos por un intervalo. Por ejemplo la función escalón de Heaviside: 

$$H(x) = 
     \begin{cases}
        0, &\quad \text{para, } x < 0 \\
        1,  &\quad\text{para. } x \ge 0 \\
     \end{cases}
$$

```python

def H(x):
  Y=np.zeros(len(x))
  for idx,x in enumerate(x): 
    if x>=0:
      Y[idx]=1
  return Y

N=1000
x=np.linspace(start=-10, stop=10, num=N)
y=H(x)

# grafica
fig, axes1=plt.subplots(nrows=1, ncols=1, figsize=(5,2))
axes1.plot(x,y)
plt.show()
```

##### funcion valor absoluto

```python

x=np.linspace(start=-10, stop=10, num=100)

def f(x):
    return np.absolute(x)

y=f(x)

# grafica
fig, axes1=plt.subplots(nrows=1, ncols=1)
axes1.plot(x,y, "b", label="|x|")
axes1.set_title(label="Función valor absoluto")
axes1.set_xlabel(xlabel="x")
axes1.set_ylabel(ylabel="y")
axes1.legend()
plt.show()
```

## Composición de funciones

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N=1000
x=np.linspace(start=-10, stop=10, num=N)

def g(x):
    return x**2

def f(x):
    return np.sin(x)

#composición de funciones
f_o_g=f(g(x))


#grafica
fig, (axes1, axes2)=plt.subplots(nrows=1, ncols=2, figsize=(12,3))
axes1.plot(x, g(x))
axes1.set_xlabel(xlabel="x")

axes2.plot(x,f_o_g)
axes2.set_xlabel(xlabel="x")
plt.show()
```
